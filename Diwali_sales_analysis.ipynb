```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import csv file

df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
## list all the column names
df.columns
```

```python
# drop the blank columns

df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
## checking for null values

df.isnull().sum()
```

```python
## Using mean imputation technique to fill the null values

df['Amount_mean'] = df['Amount'].fillna(df['Amount'].mean())
```

```python
df.drop(['Amount'], axis = 1, inplace = True)
```

```python
# changing the datatypr of Amount_median column

df['Amount_mean'] = df['Amount_mean'].astype('int')
```

```python
df['Amount_mean'].dtypes
```

```python
df.head()
```

```python
## rename the columns
df.rename(columns={'Amount_mean': 'Amount'}, inplace=True)
```

```python
df.describe()
```

## Exploratory Data Analysis

### Gender

```python
## plotting a bar chart for gender and their count
sns.countplot(x='Gender', data=df)
```

```python
## plotting a bar chart for gender and total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

From the above two graphs we can see that most of the buyers are females and even the purchasing power of females is greater than males.

## Ages

```python
## plotting a count chart for ages and their counts
sns.countplot(x='Age Group', data = df, hue = 'Gender')
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

From that graph we can see that most of the buyers are from 26-35 age group.

## State

```python
## total number of users from top 10 states

top_10_states = df.groupby('State', as_index=False)['User_ID'].nunique().sort_values(by='User_ID', ascending=False).head(10)

colors = sns.color_palette('husl', len(top_10_states))
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='User_ID', data=top_10_states, palette=colors)

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('State')
plt.ylabel('Total Users')
plt.title('Total Number of Users from Top 10 States')

plt.show()
```

```python
# total number of orders from top 10 states

top_10_states = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


colors = sns.color_palette('Set2', len(top_10_states))
plt.figure(figsize=(17, 6))
sns.barplot(x='State', y='Orders', data=top_10_states, palette=colors)

plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Total Number of Orders from Top 10 States')

plt.show()
```

```python
# total amount/sales from top 10 states

# total number of orders from top 10 states

top_10_states = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)


colors = sns.color_palette('Spectral', len(top_10_states))
plt.figure(figsize=(15, 6))
sns.barplot(x='State', y='Amount', data=top_10_states, palette=colors)

plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Amount')

plt.show()
```

## Marital Status

```python
# total count of married peoples

colors = sns.color_palette('viridis', 2)
ax = sns.countplot(x='Marital_Status', data = df, palette = colors)

for container in ax.containers:
    ax.bar_label(container)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


colors = sns.color_palette('coolwarm', 2)
plt.figure(figsize=(6,5))
sns.barplot(x='Marital_Status', y='Amount', data=sales_state, palette=colors, hue = 'Gender')

plt.xticks(rotation=45)
plt.xlabel('Marital_Status')
plt.ylabel('Amount')

plt.show()
```

## Occupation

```python
sns.set(rc={'figure.figsize':(20,6)})

colors = sns.color_palette('husl', 15)
ax = sns.countplot(x='Occupation', data = df, palette = colors)
for container in ax.containers:
    ax.bar_label(container)
```
